```python
from ipywidgets import interact, interactive, fixed, interact_manual
from functools import partial
import matplotlib.pyplot as plt
import numpy as np
```

```python
def parabole(a, b):
    x = np.linspace(-10,10)
    y = x*x*a + b*x
    plt.plot(x,y)
    return 0
```

```python
parabole(1,2)
```

Nupiešėme parabolę, dbr imkim idėkim į interact su savo pasirinktais kintamaisiais

```python
interact(parabole, a =(-1, 1, 0.1), b = (-1, 1, 0.1))
```
